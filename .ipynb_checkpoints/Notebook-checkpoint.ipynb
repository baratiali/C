### Linux terminal commands


`pwd`   : prints current directory

`ls -l`    : Shows all files in the directory(file permisions, time, ...)

`ls -a`    : Shows all files and hidden files in the directory

`ls -la`    : mix of the both

`cd /`  : get to very first dir

`cd `   : Home dir

`cd ~`  : Home dir(you can refrence with it. next command as exmple)

`cd ~ /C-Programming`

`ls ..` : shows all inside the parents folder

`ls ../..` : shows all inside the parents, parents folder

`cd ../../OneDrive` : Going directly to the folder from where you are back

`touch` : Create a file

`mv filename  ..`   : Move the file into the parent folder (instead of `..` you can type the folder name)

`mv ../Notebook.ipynb .  `  : Moving forward

`mv filename newfilename`   : Rename file

`rm filename`   : TO remove the file

`mkdir 01 02 03 `   : You can make several dir by put space between them

`echo`  : take whatever you give and put it back to terminal

`echo "Hello World > output`    : Put the Hello world in the file named output

`cat output`    : To print out the output file

`echo " Pizza >> output`    : to append Pizza in the file named output

### 03 - Taking Input from User
------
```c
#include <stdio.h>


int main()

{
    int radius;
    printf("Please insert the radius: ");// user prompt
    scanf("%i", &radius); //address of operator 
    
    printf("the given radius is = %i\n" , radius);

    return 0;


}
```
------

### 04 - Arithmetic Expressions

--------------
```c
#include <stdio.h>


int main()

{
    int radius;
    printf("Please insert the radius: ");// user prompt
    scanf("%i", &radius); //address of operator 

    double area = 3.14159 * (radius * radius);
    
    //printf("the given area is = %i\n" , area); // i is just for integer numbers
    
    printf("the  area is = %f\n" , area); // we use instead of "i", "f"
    
    return 0;


}


```
---------------

### 05 - Basic Type Casting

--------------
```c
#include <stdio.h>


int main()

{
    int eggs;
    printf("The number of eggs for the day: ");
    scanf("%i", &eggs);  

    double dozen = (double) eggs / 12 ; // dozen is 12
    // If we dont use double before eggs, the result will not be true 
    
    printf("You have %f dozen eggs.\n" , dozen);
    
    return 0;


}


```
---------------

### 06 - Working with Strings

--------------
```c

#include <stdio.h>


int main()

{
    char name[100]; //  one more than the number is allowed

    printf("What is your name: ");
    
    scanf("%s", name);// yout dont need to add "&" for chars

    printf("Hello %s \n" , name);
    
    return 0;


}


```

### 07 - Data Types

#### Conversion Characters

+ `%f`: Decimal notation (2500000)
+ `%e`: Scientific notation (2.5e4)
+ `%g`: Coputer decides
--------------
```c
#include <stdio.h>


int main()

{
    double dogs;
    scanf("%lf", &dogs); // we can use lf for all of them

    printf("%f\n%e\n%g\n" , dogs, dogs, dogs);
    
    return 0;



}


```
